#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
x = np.arange(0,100)
y=x**2
z=np.exp(x)
fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(8,8))
axs[0].plot(x,y,z)
axs[0].set_xscale("linear") 
axs[0].set_yscale("linear")
axs[1].plot(x,y,z)
axs[1].set_xscale("log") 
axs[1].set_yscale("log")
axs[0].set_title('normal scale graphic',fontsize=14,horizontalalignment='center')
axs[1].set_title('logarithmic scale graphic',fontsize=14,horizontalalignment='center')
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
x = np.arange(0,100)
y=x**2
z=np.exp(x)
fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(13,8))
axs[0].plot(x,y,color = "r", label = "y = x^2")
axs[0].plot(x,z,color = "b",label = "z = e^x")
axs[0].set_xscale("linear") 
axs[0].set_yscale("linear")
axs[1].plot(x,y,color = "g", label = "y = x^2")
axs[1].plot(x,z,color = "orange", label = "z = e^x")
axs[1].set_xscale("log") 
axs[1].set_yscale("log")
axs[0].grid(True)
axs[1].grid(True)
axs[0].legend()
axs[1].legend()
axs[0].set_title('normal scale graphic',fontsize=14,horizontalalignment='center')
axs[1].set_title('logarithmic scale graphic',fontsize=14,horizontalalignment='center')
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
month_encoded = fitbit.groupby('Months_encoded')
mean = month_encoded.mean().reset_index()
median = month_encoded.median().reset_index()
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
fig, axs = plt.subplots(ncols=1, nrows=2, figsize=(12,4))
axs[0].bar(mean['Months_encoded'],mean['Steps'])
axs[1].bar(median['Months_encoded'],median['Steps'])
plt.savefig("output.png", bbox_inches="tight")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
plots= ['Minutes Lightly Active', 'Minutes Very Active','Minutes Sedentary' ]
for i in range(3):
    plt.scatter(fitbit[plots[i]],fitbit["Steps"], marker='x',label=plots[i]+' vs Steps')
    plt.legend(bbox_to_anchor=(1.04,1), borderaxespad=0)
```
